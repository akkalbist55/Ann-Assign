```python
import numpy as np
def sigmoid(x):
    return 1 / (1 + np.exp(-x))
```

```python
def sigmoid_derivative(x):
    return sigmoid(x) * (1 - sigmoid(x))
```

```python
def forward_pass(X, W1, b1, W2, b2):
    # Calculate the output of the hidden layer
    hidden_layer_output = sigmoid(np.dot(X, W1) + b1)
    # Calculate the output of the output layer
    output_layer_output = sigmoid(np.dot(hidden_layer_output, W2) + b2)
    return hidden_layer_output, output_layer_output
```

```python
def backpropagation(X, y, W1, b1, W2, b2, learning_rate):
    # Forward pass
    hidden_layer_output, output_layer_output = forward_pass(X, W1, b1, W2, b2)
    # Calculate the error
    error = y - output_layer_output
    # Backpropagate the error through the output layer
    output_layer_error = error * sigmoid_derivative(output_layer_output)
    # Backpropagate the error through the hidden layer
    hidden_layer_error = np.dot(output_layer_error, W2.T) * sigmoid_derivative(hidden_layer_output)
    # Update the weights and biases
    W2 += np.dot(hidden_layer_output.T, output_layer_error) * learning_rate
    b2 += np.sum(output_layer_error, axis=0, keepdims=True) * learning_rate
    W1 += np.dot(X.T, hidden_layer_error) * learning_rate
    b1 += np.sum(hidden_layer_error, axis=0, keepdims=True) * learning_rate
    return W1, b1, W2, b2
```

```python
import numpy as np

# Define the sigmoid activation function and its derivative
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def sigmoid_derivative(x):
    return x * (1 - x)

# Define the input data
X = np.array([[0, 0], [0, 1], [1, 0], [1, 1]])

# Define the target data
y = np.array([[0], [1], [1], [0]])

# Define the initial weights for the neural network
np.random.seed(1)
w1 = np.random.randn(2, 2)
w2 = np.random.randn(2, 2)
w3 = np.random.randn(2, 1)

# Define the learning rate and number of iterations
learning_rate = 0.1
num_iterations = 10000

# Train the neural network using backpropagation
for i in range(num_iterations):
    # Forward propagation
    z2 = np.dot(X, w1)
    a2 = sigmoid(z2)
    z3 = np.dot(a2, w2)
    a3 = sigmoid(z3)
    z4 = np.dot(a3, w3)
    y_pred = sigmoid(z4)

    # Backward propagation
    delta4 = (y_pred - y) * sigmoid_derivative(y_pred)
    delta3 = np.dot(delta4, w3.T) * sigmoid_derivative(a3)
    delta2 = np.dot(delta3, w2.T) * sigmoid_derivative(a2)

    # Update the weights
    w3 -= learning_rate * np.dot(a3.T, delta4)
    w2 -= learning_rate * np.dot(a2.T, delta3)
    w1 -= learning_rate * np.dot(X.T, delta2)

# Print the final predictions
print(y_pred)
```
